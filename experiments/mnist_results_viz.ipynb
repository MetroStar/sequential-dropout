# visualize experiments results

```python
import json
import matplotlib.pyplot as plt
```

```python
results = json.load(open('results/window_result.json'))

```

```python
# Training losses
for window, result in zip(results['config']['windows'], results['window_results']):
    label = window[2]
    loss=result['loss']
    r = result['emb_dim']
    plt.plot(loss['train_loss'],label= label+"_train")
    plt.plot(loss['val_loss'],label = label+"_val") 
    plt.xlabel("epochs")
    plt.ylabel("Loss")
plt.legend(loc="upper right")
plt.show()
```

```python
for window, result in zip(results['config']['windows'], results['window_results']):
    label = window[2]
    avg_psnrs=result['psnr']
    r = result['emb_dim']
    plt.plot(r, avg_psnrs,label= label)
    plt.xlabel("embedding size")
    plt.ylabel("PSNR")
plt.legend(loc="lower right")
plt.show()
```
